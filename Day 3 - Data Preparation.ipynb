```python
import pandas as pd
data = pd.read_csv("C:\\Users\\DELL\\Downloads\\ML + NN by iNeuron\\Session_3\\purchase.csv")
data

# Data has missing values -- Incomplete data
```

```python
data.describe()
```

```python
# mean
data['Ball'] = data['Ball'].fillna(data['Ball'].mean())
data

# Filling mean values on NA of Ball feature
```

```python
# median
data['Bat'] = data['Bat'].fillna(data['Bat'].median())
data

# Filling median values on NA of Bat feature
```

```python
# standard deviation  
data['apple'] = data['apple'].fillna(data['apple'].std())
data

# Filling standard deviation value on NA of Apple feature
```

```python
# min  
data['Orange'] = data['Orange'].fillna(data['Orange'].min())
data

# Filling min value on NA of Orange feature
```

```python
# max  
data['Price'] = data['Price'].fillna(data['Price'].max())
data

# Filling maximum value on NA of Price feature
```

```python
# Normalization (also called, Min-Max normalization) is a scaling technique such that when it is applied the features will be rescaled so that the data will fall in the range of [0,1]
# x' = [x - min(x)]/[max(x) - min(x)]
```

```python
from pandas import read_csv
from numpy import set_printoptions
from sklearn import preprocessing
data = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv'

name = ["A","B","C","D","E","F","G","H","I"]
a = read_csv (data, names=name)
# creating header for our dataset.

a
```

```python
a.describe()
```

```python
# code for normalization

scaler = preprocessing.MinMaxScaler(feature_range=(0,1))
rescaled = scaler.fit_transform(a)
set_printoptions(precision=2) # decimal value
rescaled
```

```python
# Standardization(also called, Z-score normalization) is a scaling technique such that when it is applied the features will be rescaled such that they'll have properties of a standard normal distribution.
# z = (x-u)/sigma      u - mean, sigma - standard deviation
```

```python
# Code for Standardization

from sklearn.preprocessing import StandardScaler
data_scaler = StandardScaler().fit(a)
data_rescaled = data_scaler.transform(a)
data_rescaled
```

```python
# Binarization - converts into binary w.r.t. threshold
# 0 for <= threshold, 1 for > threshold
```

```python
# code for Binarization

from sklearn.preprocessing import Binarizer
binary = Binarizer(threshold=0.5)
binary1 = binary.transform(a)
binary1
```
