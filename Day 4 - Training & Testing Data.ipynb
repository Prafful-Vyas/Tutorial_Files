```python
import pandas as pd
df = pd.read_csv("C:\\Users\\DELL\\Downloads\\ML + NN by iNeuron\\Session_2\\iris_data.csv")
df.head()
```

```python
X = df[['sepal_length','sepal_width','petal_length','petal_width']]
Y = df['species']

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=5)
# train_test_split(Ind. var, Dep. var,test_size/train_size,random_state)
# test_size -- 0.2 means 20% is used to test data
# random state -- data is shuffled randomly and states are created to reduce biasness 1,2,3 is fine for beginners

X_train
```
